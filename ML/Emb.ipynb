## EMb

```python
import torch
import numpy as np
import pandas as pd
from transformers import AutoTokenizer, AutoModel
from tqdm import tqdm
import matplotlib.pyplot as plt

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

molformer_model_name = "ibm-research/MoLFormer-XL-both-10pct"
molformer_tokenizer = AutoTokenizer.from_pretrained(molformer_model_name, trust_remote_code=True)
molformer_model = AutoModel.from_pretrained(molformer_model_name, trust_remote_code=True, deterministic_eval=True).to(device).eval()

def get_molformer_embedding(smiles: str):
    inputs = molformer_tokenizer([smiles], padding=True, truncation=True, return_tensors="pt").to(device)
    with torch.no_grad():
        outputs = molformer_model(**inputs)
    return outputs.pooler_output.cpu().numpy().squeeze()

seyonec_model_name = "seyonec/PubChem10M_SMILES_BPE_450k"
seyonec_tokenizer = AutoTokenizer.from_pretrained(seyonec_model_name)
seyonec_model = AutoModel.from_pretrained(seyonec_model_name).to(device).eval()

def get_seyonec_embedding(smiles: str):
    inputs = seyonec_tokenizer(smiles, return_tensors="pt", padding=True, truncation=True).to(device)
    with torch.no_grad():
        outputs = seyonec_model(**inputs)
    emb = outputs.last_hidden_state.mean(dim=1)
    return emb.cpu().numpy().squeeze()

def embed_series(series, embed_func, desc=""):
    embeddings = [embed_func(smi) for smi in tqdm(series, desc=desc, disable=True)]
    return np.vstack(embeddings)
```

```python
train_molformer = embed_series(train_data['Canonical_Smiles'], get_molformer_embedding, "MoLFormer Embedding (train)")
test_molformer = embed_series(test_data['Canonical_Smiles'], get_molformer_embedding, "MoLFormer Embedding (test)")
train_seyonec = embed_series(train_data['Canonical_Smiles'], get_seyonec_embedding, "Seyonec Embedding (train)")
test_seyonec = embed_series(test_data['Canonical_Smiles'], get_seyonec_embedding, "Seyonec Embedding (test)")
train_embeddings = np.concatenate([train_molformer, train_seyonec], axis=1)
test_embeddings = np.concatenate([test_molformer, test_seyonec], axis=1)
embedding_dim = train_embeddings.shape[1]
col_names = [f'smiles_emb_{i}' for i in range(embedding_dim)]
train_emb_df = pd.DataFrame(train_embeddings, columns=col_names)
test_emb_df = pd.DataFrame(test_embeddings, columns=col_names)
train_data_em = pd.concat([train_data, train_emb_df], axis=1)
test_data_em = pd.concat([test_data, test_emb_df], axis=1)
train_data_em.to_parquet('train_emb_final.parquet', index=False)
test_data_em.to_parquet('test_emb_final.parquet', index=False)
```
